## 2. CSV

```python
import pandas as pd
data = pd.read_csv('/content/drive/My Drive/epidemic-curve.csv', header=None)
data.head()
```

```python
#Reading csv file
f = open('/content/drive/My Drive/epidemic-curve.csv', 'r')
lines = f.readlines()
lines2 = []

for line in lines:

  #.strip allows you to remove the whitespace from the beginning or end of the string
  line = line.strip()
  lines2.append(line)

print(lines2)
f.close()

```

```python
#Tokenisation
ep_curve = []
for line in lines2:
  ep ={}

  #.split allows you to split a string into a list, default seperator is any whitespace
  tokens = line.split(',')
  date = tokens[0]
  type_ep = tokens[1]
  value = int(tokens[2])

  #Set the key-value pairs in our dictionary ep

  ep['date'] = date
  ep['type_ep'] = type_ep
  ep['value'] = value
  ep_curve.append(ep)

print(ep_curve)
```

## 3. Statistics


### Q) Find the average value of the number of caxes in the world

```python
#1. For loop the whole list 
total = 0

for ep in ep_curve:
  value = ep['value']
  # a += b basically means a = a + b
  total += value

#len() function is a built-in function of python, returns the length of an object, can be different types
print('Average value of epidemic curve is ')
print(total/len(ep_curve))
```

### Q) Compute the mean value of cases each type: Imported, Local

```python
from collections import Counter
```

```python
#Example
#Counter('Aaron')
```

```python
mean_averages = {}
counts = {}

for ep in ep_curve:
  type_of_case = ep['type_ep']
  value = ep['value']

  #If the current part is not recorded yet
  if (type_of_case not in mean_averages):
    mean_averages[type_of_case] = value
    counts[type_of_case] = 1

  else:
    #Add the current epidemic value into the count
    mean_averages[type_of_case] += value
    counts[type_of_case] += 1

print(mean_averages)
print(counts)
average_value = Counter({key : mean_averages[key] // counts[key] for key in mean_averages})

print('The average count for each type of case is : ' + str(dict(average_value)))
```
