## What is a class?
<p>A class is like an object constructor, or a blueprint for creating objects</p>
<p> Classes provide a means of bundling data and functionality together</p>
<p> Creating a new class creates a new type of object, allowing new instances of that type to be made.</p>
<p> Each class instance can have attributes attaches to it for maintaining its state</p>
<p> Lets go into out hands-on task for creating a class object!</p>

<p>For example: for a class 'Student' and one of the attribute is 'year_of_study', to access this attribute -> 'Student.year_of_study' would be the code to perform that function</p>

```python
class Student:
  year_of_study = 3
```

```python
Student.year_of_study
```

```python
Daniel = Student()
```

```python
print(Daniel.year_of_study)
```

```python
class Student:
  #class attribute
  types = "university student"

  # instance attribute
  def __init__(self, name, yos, cca):
    self.name = name
    self.yos = yos
    self.cca = cca
```

```python
# Instantiate the Student class
Chandler = Student("Chandler", 3, "DAC")
Ross = Student("Ross", 3, "IONS")
Racheal = Student("Racheal", 1, "Dragon Boating")
```

```python
# Access the class attributes
print("Chandler is a {}".format(Chandler.__class__.types))
print("Ross is a {}".format(Ross.__class__.types))
print("Racheal is a {}".format(Racheal.__class__.types))
```

```python
print("{} is in his Year {} of study pursuing {} as his CCA".format(Chandler.name, Chandler.yos, Chandler.cca))
print("{} is in his Year {} of study pursuing {} as his CCA".format(Ross.name, Ross.yos, Ross.cca))
print("{} is in her Year {} of study pursuing {} as her CCA".format(Racheal.name, Racheal.yos, Racheal.cca))
```
