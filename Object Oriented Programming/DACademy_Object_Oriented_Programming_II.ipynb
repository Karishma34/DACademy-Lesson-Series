## Encapsulation
* describes the idea of wrapping data and the methods that work on data within one unit
* puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data
* Object's variable can only be changed by an object's method -> known as private variables

Real life example:
* Hiding data from one department of the company to the other departments for confidentiality purposes 

```python
class Book: 
  def __init__(self, title, quantity,author,price):
    self.title = title
    self.quantity = quantity
    self.author = author
    #Private
    self.__price = price
    self.__discount = None
  
  def set_discount(self, discount):
    self.__discount = discount
  
  def get_price(self):
    if self.__discount:
      return self.__price * (1-self.__discount)
    return self.__price
  
  def __repr__(self):
    return f"Book: {self.title}, Quantity: {self.quantity}, Author: {self.author}, Price: {self.get_price()}"
```

```python
# Instantiate the objects
# You do not include the private attributes in which would be the price
singular_books = Book('The Lost Art of Doing Nothing', 1, 'Maartje Willems', 17.99)
books_in_bulk = Book('The Fault in Our Stars', 25, 'John Green', 43.73)
books_in_bulk.set_discount(0.15)

#Print out the prices of singular book
print(singular_books.get_price())
#Print out the prices of books in bulk 
print(books_in_bulk.get_price())

#Printing out entire objects
print(singular_books)
print(books_in_bulk)
```

## Inheritance

Definition: 
Inheritance is the capability of one class to derive or inherit the properties from another class
* The class has the ability to inherit methods and characteristics from another class
* The child class is the one that inherits.
* Parent class is the class from which the methods are inherited from
* Used to reduce computational time
* Simply inherit the properties of one class into the other

```python
#Parent Class
class Student(object):

  #__init__ is know as the constructor
  def __init__(self, name, idnumber):
    self.name = name
    self.idnumber = idnumber

  def display(self):
    print(self.name)
    print(self.idnumber)

#Child class
class Administration(Student):
  def __init__(self, name, idnumber, grades_gpa, position):
    self.grades_gpa = grades_gpa
    self.position = position

    #invoking the __init__ of the parent class
    Student.__init__(self, name, idnumber)

#Creating an object variable or an instance
a = Administration('Chandler', 886012, 4.0, "Head Prefect")

#Calling a function of the class Student using its instance
a.display()
```

## Task 1:
* Define an instance method that calculates the average gpa of a student. Call it average_gpa
* Define the "str" function that returns the student's name, university and major

```python
class Student:
  #Init function
  def __init__(self, user_name, university, major, gpas):
    self.user_name = user_name
    self.university = university
    self.major = major
    self.gpas = gpas
  
  #Function to calculate the gpa of a person
  def average_gpa(self):
    #1. Define a total variable and set it to 0
    total = 0.0
    #2. Loop through the gpas instant variables
    for gpa in self.gpas:
      #3. Add the gpa to the total
      total = total + gpa
    #4. Calculate the mean by dividing the total to the number of elements 
    num_elems = len(self.gpas)
    mean_gpa = total/num_elems

    return mean_gpa

  #String function
  def __str__(self):
    return f'Name: {self.user_name}\nUniversity:{self.university}\nMajor:{self.major}'
  
```

### Instatiating the object

```python
student1 = Student("Chandramohan Amudha Karishma", "UOL", "Computer Science", [89,92,97,88])
average = student1.average_gpa()
print(f'Average GPA of the student is {average}')
```

* Print out the average gpa of each student

```python
#Read line by line
f = open('/content/drive/MyDrive/Copy of student.csv', 'r')
lines = f.readlines()
f.close
```

```python
#Store student data objects
students = []
#Process each line
for line in lines:
  # Strip all white space characters from the line (white spaces = empty spaces, new line character, tab, ...)
  line = line.strip()

  #Tokenize the string
  tokens = line.split(',')
  
  name = tokens[0]
  university = tokens[1]
  major = tokens[2]

  #arrange index
  #Take tokens from index 3 till the end
  gpas = tokens[3:]
  int_gpas = []

  #convert strings to integers
  #Parse all gpas into int
  #List comprehension
  #gpas = [int(x) for x in gpas]
  for gpa in gpas:
    gpa = int(gpa)
    int_gpas.append(gpa)
    
  student = Student(name, university, major, int_gpas)
  students.append(student)
```

```python
for student in students:
  average_gpa = student.average_gpa()
  print(f'Name : {student.user_name}, Average GPA = {average_gpa}')
```
